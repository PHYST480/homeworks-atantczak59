```python
# Problem 3
# Produce the "trace" plot for the clear/cloudy day problem from Inference2.ipynb. 
# Then plot the histogram of that trace, throwing away steps in the trace that could 
# be considered part of the burn-in. Your plots should look like the ones from class (or nicer!). 
# Confirm that you get the same answer for day  N as day  N+1
# and it doesn't matter whether is was clear or cloudy on the day that you started.

# This is a re-do of Problem 3 because I'm missing some things. 

Sunny = 1
Cloudy = 0

N = 10000

def sky(N, start_day):
    today = start_day
    conditions = []
    days = []
    day_count = 1
    Sunny_P_Trace = []
    
    for i in range(N):
        prob = np.random.random(1)
        if today == Sunny:
            if prob <= 0.5:
                tomorrow = Sunny
            else:
                tomorrow = Cloudy
        if today == Cloudy:
            if prob <= 0.9:
                tomorrow = Sunny
            else:
                tomorrow = Cloudy
                
        today = tomorrow
        conditions.append(today)
        days.append(day_count)
        conditions_sum = np.sum(conditions)
        Sunny_P = 1.*(conditions_sum)/day_count
        Sunny_P_Trace.append(Sunny_P)
        day_count += 1
    return Sunny_P_Trace, days
Sunny_P_Trace, daily = sky(N, Sunny)
```

```python
Sunny_P_Trace, daily = sky(N, Sunny)
Sunny_P_Trace_2, daily = sky(N, Cloudy)
plt.figure(figsize = (6,6))
plt.title('Sunny vs. Cloudy Weather Probability')
plt.xlabel('Days')
plt.ylabel('Probability of having a Sunny Day')
plt.plot(daily, Sunny_P_Trace, label='Start on a Sunny Day')
plt.plot(daily, Sunny_P_Trace_2, label='Start on a Cloudy Day')
plt.legend()
plt.show()
```

```python
# Does the probability of N+1 equal that of 1?

Sunny_P_Trace_NS, daily = sky(N+1, Sunny)
Sunny_P_Trace_NC, daily = sky(N+1, Cloudy)

print('Probability of having a sunny day on the Nth day: {:.4f}, '.format(Sunny_P_Trace[-1]))
print('Probability of having a cloudy day on the Nth day: {:.4f}, '.format(Sunny_P_Trace_2[-1]))
print('Probability of having a sunny day on the (N+1)th day: {:.4f}, '.format(Sunny_P_Trace_NS[-1]))
print('Probability of having a cloudy day on the (N+1)th day: {:.4f}, '.format(Sunny_P_Trace_NC[-1]))
```

```python
plt.hist(Sunny_P_Trace, bins=100)
plt.title('Probability of having a Sunny Day when starting on a Sunny Day')
plt.xlabel('Probability of a Sunny Day')
plt.ylabel('# of Occurences')
plt.xlim(0.5,0.7)
plt.show()

plt.figure()
plt.hist(Sunny_P_Trace_2, bins = 100)
plt.title('Probability of having a Sunny Day when starting on a Cloudy Day')
plt.xlabel('Probability of a Sunny Day')
plt.ylabel('# of Occurences')
plt.xlim(0.6,0.7)
plt.show()
```
